```python
from vernam_cipher import vernam_encrypt, vernam_decrypt

message = "Tajna wiadomość"
cipher, key = vernam_encrypt(message)
decrypted = vernam_decrypt(cipher, key)
print(f"Wiadomość: {message}")
print(f"Zaszyfrowane: {cipher}")
print(f"Klucz: {key}")
print(f"Odszyfrowane: {decrypted}")
```
